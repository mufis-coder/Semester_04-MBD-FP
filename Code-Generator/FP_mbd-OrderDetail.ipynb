```python
order_ids = list(order_detail['order_id'])
product_ids = list(order_detail['product_id'])
```

```python
cekd = list(tuple(zip(order_ids, product_ids)))
len(cekd)
```

```python
cekd[:5]
```

```python
def cek(arg):
  if (arg in cekd):
    return True
  return False
```

```python
cek((10248, 91))
```

```python
QRY = []
for i in range(0, 500000):
  order_id = 10248 + i + 1
  product_id = random.choice(prdID)
  cekt = (order_id, product_id)
  while(True):
    if (cek(cekt)):
      product_id = random.choice(prdID)
      cekt = (order_id, product_id)
    else:
      break
  cekd.append(cekt)
  unit_price	= round(random.uniform(2.0, 263.5), 2)
  quantity = random.choice(range(1, 130))
  discount = random.choice(disc)
  Query = "INSERT INTO order_details VALUES (" + str(order_id) + ", " + str(product_id) + ", " + str(unit_price)
  Query += ", " + str(quantity) + ", " + str(discount) + ");"
  QRY.append(Query)
  if(i%50000 == 0):
    print(i)
```
