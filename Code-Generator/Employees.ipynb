```python
def conv(data):
  if(data<10):
    return "0" + str(data)
  else:
    return str(data)
```

```python
def datef(date):
  date1 = date
  dated = datetime.strptime(date, '%Y-%m-%d')
  dated2 = dated
  dated2 += timedelta(days=random.choice([9300, 11000, 13000]))
  # dated3 += timedelta(days=random.choice(range(1, 15)))

  date2 = str(dated2.year) + "-" + conv(dated2.month) + "-" + conv(dated2.day)
  # date3 = str(dated3.year) + "-" + conv(dated3.month) + "-" + conv(dated3.day)
  return date1, date2
```
