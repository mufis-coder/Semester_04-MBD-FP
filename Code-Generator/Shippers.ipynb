```python
shipper_id = []
company_name = []
phone = []
QRY = []
for i in range(0, 18) :
  shipper_id = 6+(i+1)
  company_name = fake.company()
  phone = fake.phone_number()
  Query = "INSERT INTO shippers VALUES (" + str(shipper_id) + ", " + "'" + str(company_name) + "'" + ", " + "'" + str(phone) + "'" + ");"
  QRY.append(Query)
```
