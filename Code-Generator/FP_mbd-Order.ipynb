```python
from faker import Faker
fake = Faker()
from datetime import datetime
from calendar import monthrange
from datetime import datetime, timedelta
import decimal
```

```python
def conv(data):
  if(data<10):
    return "0" + str(data)
  else:
    return str(data)
```

```python
def datef(date):
  date1 = date
  dated = datetime.strptime(date, '%Y-%m-%d')
  dated2, dated3 = dated, dated
  dated2 += timedelta(days=random.choice([30, 60, 90]))
  dated3 += timedelta(days=random.choice(range(1, 15)))

  date2 = str(dated2.year) + "-" + conv(dated2.month) + "-" + conv(dated2.day)
  date3 = str(dated3.year) + "-" + conv(dated3.month) + "-" + conv(dated3.day)
  return date1, date2, date3
```

```python
customer = cus.copy()
cusidP = list(customer['customer_id'])
cusnameP = list(customer['company_name'])
dictcus = dict(zip(cusidP, cusnameP))
```

```python
def name_kapal(id):
  return dictcus[id]
```

```python
def sql(inpt):
  if(type(inpt)==str):
    return "'{}'".format(inpt.replace("'", ""))
  else:
    return str(inpt)
```

```python
order_id = []
customer_id = []
employee_id = []
order_date = []
required_date = []
shipped_date = []
ship_via = []
freight = []
ship_name = []
ship_address = []
ship_city = []
ship_region = []
ship_postal_code = []
ship_country = []
for i in range(0, 500000):
   order_id.append(11077 + (i + 1))
   cusid = random.choice(list_cusid)
   customer_id.append(cusid)
   employee_id.append(random.choice(range(1, 10)))
   date1, date2, date3 = datef(fake.date())
   order_date.append(date1)
   required_date.append(date2)
   shipped_date.append(date3)
   ship_via.append(random.choice(range(1, 4)))
   freight.append(round(random.uniform(0.02, 1007.64), 2))
   ship_name.append(name_kapal(cusid))
   addres = fake
   ship_address.append(addres.address())
   ship_city.append(addres.city())
   ship_region.append(random.choice(shipr))
   ship_postal_code.append(addres.postcode())
   ship_country.append(addres.country())
```

```python
df = pd.DataFrame(index=range(0, 500000))
```

```python
df['order_id'] = order_id
df['customer_id'] = customer_id
df['employee_id'] = employee_id
df['order_date'] = order_date
df['required_date'] = required_date
df['shipped_date'] = shipped_date
df['ship_via'] = ship_via
df['freight'] = freight
df['ship_name'] = ship_name
df['ship_address'] = ship_address
df['ship_city'] = ship_city
df['ship_region'] = ship_region
df['ship_postal_code'] = ship_postal_code
df['ship_country'] = ship_country
```

```python
df['order_id'] = df['order_id'].astype(int)
df['employee_id'] = df['employee_id'].astype(int)
df['ship_via'] = df['ship_via'].astype(int)
```
