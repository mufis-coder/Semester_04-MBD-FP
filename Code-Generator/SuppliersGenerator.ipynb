```python
from faker import Faker
import random
fake = Faker()
```

```python
supplier_id = []
company_name = []
contact_name = []
contact_title = []
address = []
city = []
region = []
postal_code = []
country = []
phone = []
fax = []
homepage = []
QRY = []
for i in range(0, 20):
  supplier_id.append(30 + (i + 1))
  company_name.append(fake.swift())
  contact_name.append(fake.name())
  contact_title.append(random.choice(sup_df['contact_title']))
  address.append(fake.address())
  city.append(fake.city())
  region.append(random.choice(sup_df['region']))
  postal_code.append(fake.postcode())
  country.append(fake.country())
  phone.append(fake.phone_number())
  fax.append(fake.phone_number())
  homepage.append('#'+fake.swift())
  Query = "INSERT INTO suppliers VALUES (" + str(supplier_id[i]) + ", " + "'" + str(company_name[i]) + "'" + ", " + "'" + str(contact_name[i]) + "'"
  Query += ", " + "'" + str(contact_title[i]) + "'" + ", " + "'" + str(address[i]) + "'" + ", " + "'" + str(city[i]) + "'" + ", " + "'" + str(region[i]) + "'"
  Query += ", " + str(postal_code[i]) + ", " + "'" + str(country[i]) + "'" + ", " + "'" + str(phone[i]) + "'" + ", " + "'" + str(fax[i]) + "'" + ", " + "'" + str(homepage[i]) + "'" +");"
  QRY.append(Query)
```

```python
QRY
```

```python
textfile = open('/content/drive/MyDrive/FP-MBD/Hasil/hasil-suppliers.sql', 'w')
for element in QRY:
    textfile.write(element + "\n")
textfile.close()
```
