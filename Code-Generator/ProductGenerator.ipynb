```python
product_df.columns
from faker import Faker
import random
fake = Faker()
```

```python
product_id = []
product_name = []
supplier_id = []
category_id = []
quantity_per_unit = []
unit_price = [] 
units_in_stock = []
units_on_order = []
reorder_level = [] 
discontinued = []
QRY = []
for i in range(0, 20):
  product_id.append(78 + (i + 1))
  product_name.append(fake.swift())
  supplier_id.append(random.choice(range(1,30)))
  category_id.append(random.choice(range(1,9)))
  quantity_per_unit.append(random.choice(product_df['quantity_per_unit']))
  unit_price.append(round(random.uniform(263.50, 2.54), 2))
  units_in_stock.append(random.choice(range(0, 125)))
  units_on_order.append(random.choice(range(0, 100)))
  reorder_level.append(random.choice(range(0, 31)))
  discontinued.append(random.choice(range(0, 2)))
  Query = "INSERT INTO products VALUES (" + str(product_id[i]) + ", " + "'" + str(product_name[i]) + "'" + ", " + str(supplier_id[i])
  Query += ", " + str(category_id[i]) + ", " + "'" + str(quantity_per_unit[i]) + "'" + ", " + str(unit_price[i]) + ", " + str(units_in_stock[i])
  Query += ", " + str(units_on_order[i]) + ", " + str(reorder_level[i]) + ", " + str(discontinued[i]) +");"
  QRY.append(Query)
```

```python
QRY
```

```python
textfile = open('/content/drive/MyDrive/FP-MBD/Hasil/hasil-products.sql', 'w')
for element in QRY:
    textfile.write(element + "\n")
textfile.close()
```
