```python
from faker import Faker
import random
fake = Faker()
```

```python
territory_id = []
territory_description = []
region_id = []
QRY = []
for i in range(0, 152) :
  territory_id = 53+(i+1)
  territory_description = fake.city()
  region_id = random.choice(range(1,5))
  Query = "INSERT INTO territories VALUES (" + str(territory_id) + ", " + "'" + str(territory_description) + "'" + ", " + str(region_id) + ");"
  QRY.append(Query)
```
